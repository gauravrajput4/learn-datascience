## Numpy Indexing and slicing of vectors

```python
import numpy as np
```

```python
arr=np.arange(11,21)
arr
```

```python
arr[6]
```

```python
arr[0:5]
```

```python
arr[4:9:3]
```

## Numpy Indexing and slicing of matrix

```python
arr=np.arange(1,31).reshape(6,5)
arr
```

```python
arr[0]
```

```python
arr[0,0]
```

```python
arr[0:2,1:3]
```

```python
arr[3:,3:]
```

```python
arr[:,2]
```

## Boolean Indexing

```python
arr=np.arange(11,21)
arr
```

```python
bool_index=arr%2==0
```

```python
arr=arr[bool_index]
arr
```
