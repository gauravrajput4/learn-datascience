```python
import numpy as np
```

## Arithmetic Operation

```python
a1=np.array([1,2,3,4,5])
a2=np.array([6,7,8,9,10])
```

```python
print(a1,a2)
```

```python
# size must be same
sum=a1+a2
sum
```

```python
sub=a1-a2
sub
```

```python
mul=a1*a2
mul
```

```python
div=a1/a2
div
```

```python
power=a1**a2
power
```

## Broadcasting

```python
l=[10,20,30,40]
arr=np.array(l)
arr
```

```python
for i in l:
    print(i+10)
```

```python
arr+10
```

# Advance Array Manipulation

## Stacking Array

```python
a=np.array([1,2,3,4])
b=np.array([5,6,7,8])
```

```python
a
```

```python
b
```

```python
# vertical stack
np.vstack((a,b))
```

```python
## Horizontal stack
np.hstack((a,b))
```

```python
np.column_stack((a,b))
```
