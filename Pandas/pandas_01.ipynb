## Pandas Series
### A series is a one-deminsional labeled array capable of holding any data type. The axis label are collectively called the index

## Creating Series
### There are multiple ways to create a pandas series

```python
import numpy as np
import pandas as pd
```

```python
labels=['a','b','c']
my_list=[10,20,30]
my_arr=np.array([10,20,30])
dicts={1:10,2:20,3:30}
```

```python
pd.Series(my_list)
```

```python
pd.Series(my_list,index=labels)
```

```python
pd.Series(my_arr)
```

```python
pd.Series(dicts)
```
