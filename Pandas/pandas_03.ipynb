```python
import numpy as np
import pandas as pd
```

### Finding missing data

```python
data={
    'A':[1,2,np.nan,4,5],
    'B':[np.nan,2,3,4,5],
    'C':[1,2,3,np.nan,np.nan],
    'D':[1,np.nan,np.nan,np.nan,5]
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

### Removing Missing data

```python
df
```

```python
df.dropna() # It is remove those rows whose have any nan value in the rows.
```

```python
df.dropna(thresh=3)
```

### Filling missing data

```python
df
```

```python
df.fillna(0)
```

```python
values={'A':0,'B':100,'C':300,'D':400}
```

```python
df.fillna(value=values)
```

```python
df
```

```python
df.fillna(df.mean())
```
