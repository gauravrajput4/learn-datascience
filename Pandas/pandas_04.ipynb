```python
import numpy as np
import pandas as pd
```

### Merging 3 dataframes

```python
employees=pd.DataFrame({
    'emp_id':[1,2,3,4,5],
    'name':['Akhil','sahil','Peter','Linda','Bob'],
    'department':['IT','IT','HR','Finance','HR'],
})

salaries=pd.DataFrame({
    'emp_id':[1,2,3,6,7],
    'salary':[80000,85000,65000,70000,75000],
    'bonus':[10000,5000,7000,4500,4000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries,on='emp_id')
```

```python
pd.merge(employees,salaries,on='emp_id',how='inner')
```

```python
pd.merge(employees,salaries,on='emp_id',how='outer')
```

```python
pd.merge(employees,salaries,on='emp_id',how='left')
```

```python
pd.merge(employees,salaries,on='emp_id',how='right')
```

### Concatination of 2 dataframes

```python
df1=pd.DataFrame({
    'A':['A0','A1','A2'],
    'B':['B0','B1','B2'],
    'C':['C0','C1','C2']
})

df2=pd.DataFrame({
    'A':['A3','A4','A5'],
    'B':['B3','B4','B5'],
    'C':['C3','C4','C5']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```

### Joining 2 dataframes

```python
df1=pd.DataFrame({
    'name':['Alice','Bob','Charle']
},index=[1,2,3])

df2=pd.DataFrame({
    'score':[85,90,75]
},index=[2,3,4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
df2.join(df1)
```
