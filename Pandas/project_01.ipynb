## Make a new columns for episode count



```python
def extract_episode(txt):
    check=False
    data=''
    for i in txt:
        if i=='(':
            check=True
            continue
        if check ==True:
            if i==')':
                break
            data+=i
    return data
        
            
```

```python
extract_episode(df.loc[4]['Title'])
```

```python
df['Total Episodes']=df['Title'].apply(extract_episode)
```

```python
df.head()
```

```python
df['Total Episodes']=df['Total Episodes'].str.replace('eps','')
```

```python
df['Total Episodes']=df['Total Episodes'].astype(int)
```

## Make a new Columns for time stamp

```python
df.loc[0]['Title']
```

```python
def extract_time(txt):
    data=''
    for i in range(0,len(txt)):
        if txt[i]==')':
            for j in range(0,19):
                i+=1
                data=data+txt[i]
            return data
```

```python
extract_time(df.loc[0]['Title'])
```

```python
df['Time Stamp']=df['Title'].apply(extract_time)
```

```python
from dateutil.relativedelta import relativedelta
from datetime import datetime
```

```python
def claculate_total_months(period):
    try:
        start_str,end_str=period.split(' - ')
        start_date=datetime.strptime(start_str,'%b %Y')
        end_date=datetime.strptime(end_str,'%b %Y')
        r = relativedelta(end_date, start_date)
        return r.years*12+r.months+1
    except:
        return None
```

```python
df['Months']=df['Time Stamp'].apply(claculate_total_months)
```
