```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### We use distribution plots to visualize the distribution of quantitative data.
### Let's discuss some plots that allows us to visualize the distribution of a data set. These plots are:
<ul>
    <li>Displot/Histplot</li>
    <li>joinplot</li>
    <li>pairplot</li>
    <li>rugplot</li>
</ul>

```python
df=sns.load_dataset('tips')
```

```python
df['size'].unique()
```

```python
df
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins=20,kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins=20,kde=True)
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```

```python
# Pair plot
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='size',palette='rainbow')
```

```python
# rugplot
sns.rugplot(df)
```
