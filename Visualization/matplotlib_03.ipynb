```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Categorical Data Plots
## Now lets discuss using seaborn to plot categorical data! There are a few main plot types for this:
<ul>
    <li>boxplot</li>
    <li>violinplot</li>
    <li>stripplot</li>
    <li>swarmplot</li>
    <li>barplot</li>
    <li>countplot</li>
</ul>

### <b>Let's go through examples of each</b>

```python
df =sns.load_dataset('tips')
```

```python
df
```

```python
sns.countplot(x=df['sex'],hue=df['smoker'])
```

```python
sns.barplot(x=df['sex'],y=df['total_bill'])
```

### you can change the estimator object to your own functions, that convert a vector to a scaler 
### Mean (default): np.mean
### Median : np.median
### sum= np.sum
### count= len( to count the number of observation)
### standard deviation: np.std
### minimum: np.min
### maximum: np.max

```python
sns.barplot(x=df['sex'],y=df['tip'],estimator=sum)
```

```python
sns.barplot(x=df['sex'],y=df['total_bill'],estimator=sum)
```

```python
# boxplot
sns.boxplot(x='tip',y='day',data=df,palette='rainbow')
```

### boxplot and violinplots are used to shown the distribution of categirical data. A box plot (or box-and-whisker plot) shown the distriution of quantitative data in a way that facilitates camparisons between variable ir acros levels of a categorical variable. The box shows the quartiles of the datasets while the whiskers extend to show the rest of the distribution, except for points that are determined to be "outliers" using a method that is a function of the inter-quartile range.

```python
sns.violinplot(x='day',y='total_bill',data=df,)
```

## Stripplot and swarnplot
### The stripplot will draw a scatterplot where one variable is categorical. A strip plot can be drawn on its own, but it is also a good complements to a box or violin plot in cases where you want to show alll observation  along with sokme representation of this underlying distribution

### The swarnplot is similiar to stripplot , but the points are adjusted (only the categorical axis) so that they don't overlap. This gives  a better representions of the distribution of values although it does not scale as well to large numbers of observation (both in terms of the ablity to show all the points and in terms of the computation needed to arange them).
