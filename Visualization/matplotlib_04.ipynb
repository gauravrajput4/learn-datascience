# <center>Matrix Plot</center>

```python
import numpy as np
import matplotlib.pyplot as ply
import seaborn as sns
```

```python
flights=sns.load_dataset('flights')
```

```python
flights
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

# HeatMap

```python
tips_corr=tips[['total_bill','tip','size']]
```

```python
tips_corr
```

```python
# co-relation
tips_corr.corr()
```

```python
sns.heatmap(tips_corr.corr(),annot=True)
```

# Cluster Map

```python
sns.clustermap(tips_corr.corr())
```

# Pivot table Heat map

```python
flights
```

```python
pvtflights=flights.pivot_table(values='passengers',index='month',columns='year')
```

```python
pvtflights
```

```python
sns.heatmap(pvtflights)
```

# Regression Plot

## lmplot()
### regression plot is used in machine learning

```python
tips
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip', hue='sex')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex',palette='coolwarm')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex',palette='coolwarm',markers=['o','v'],scatter_kws={'s':100})
```
