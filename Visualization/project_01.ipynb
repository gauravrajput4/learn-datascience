## which team won the most matches?

```python
match_wins=df['match_winner'].value_counts()
sns.barplot(y=match_wins.index,x=match_wins.values,palette='viridis')
plt.title('Most match win by team ')
```

## Toss Decision Trends

```python
sns.countplot(x=df['toss_decision'],palette='viridis')
plt.title('Toss Decision Trends')
```

## Toss Winner vs Match Winner

```python
count=df[df['toss_winner']==df['match_winner']]['match_id'].count()
percentage=((count*100)/df.shape[0]).round(2)
```

##  10 Best bowling figures

## Which Player had the highest individual score?

```python
top_players=df.sort_values('highscore',ascending=False).head()[['top_scorer','highscore']]
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(
    data=top_players,
    x='top_scorer',
    y='highscore'
)
plt.title('Top 5 High Scorers')
plt.xlabel('Top Scorer')
plt.ylabel('High Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
